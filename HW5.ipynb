```python
#1
def Max_of_two(a,b):
    if a > b:
        return a
    return b
a , b = map(int, input().split())
print(Max_of_two(a,b))
```

```python
#2
def swap(a, b):
    a, b = b , a
    print(a, b)
a, b = map(int, input().split())
swap(a, b)
```

```python
#3
def isprime(a):
    if a < 2:
        return False
    else:
        for i in range(2, a):
            if a % i == 0:
                return False
        return True

n = int(input())
isprime(n)
```

```python
#4
def perfect_num(n):
    sum = 0
    for i in range(1, n):
        if n % i == 0:
            sum += i
    if sum == n:
        return True
    return False

n = int(input())
perfect_num(n)
```

```python
#5
def position(a, b):
    count = 1
    for x in a:
        if x != b:
            count += 1
        elif x == b:
            return count
    return -1

a = list(map(int, input().split()))
b = int(input())
print(position(a, b))
    
    
```

```python
#6
def giaithua(n):
    product = 1
    for i in range(1, n + 1):
        product *= i
    return product

n = int(input())
giaithua(n)
```

```python
#7
def calculator(a, b, c):
    if b == '+':
        return a + c
    if b == '-':
        return a - c
    if b == '*':
        return a*c
    if b == '/':
        return a / c
    
a, b, c = input().split()
a = int(a)
c = int(c)
calculator(a,b,c)
    
    
```

```python
#8
def hamming_distance(a,b):
    xor_result = a ^ b
    return bin(xor_result).count('1')
a, b = map(int, input().split())
print(hamming_distance(a,b))
```

```python
#9
def sumdigit(n):
    sum = 0
    while n > 0:
        sum += n % 10
        n = n // 10
    return sum
n = int(input())
sumdigit(n)
```

```python
#10
def isIsomorphism(a,b):
  a2b = {}
  b2a = {}
  for x , y in zip(a,b):
    if x in a2b and a2b[x] != y:
      return False
    if y in b2a and b2a[y] != x:
      return False
    a2b[x] = y
    b2a[y] = x
  return True

a, b = map(str, input().split())
print(isIsomorphism(a,b))
  
    
```
