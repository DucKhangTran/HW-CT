```python
#3
def tong_bp(a):
    sum = 0
    for i in a:
        sum += int(i) ** 2
    return sum

def happy_num(n):
    s = set()
    while n != 1:
        if n in s:
            return False
        else:
            s.add(n)
            n = tong_bp(str(n))
            
    return True

n = input()
try:
    if happy_num(n):
        print('YES')
    else:
        print('NO')
except:
    print('ValueError')


    
    
```

```python
#8
def file(a):
    n = len(a) - 1
    if a[n] == 'txt' or a[n] == 'zip':
        return True
    return False

a = input().strip()
a = a.lower().split('.')
try:
    if file(a):
        print('Doc file thanh cong')
    else:
        print('File khong hop le')
except IndexError:
    print('File khong hop le')
```

```python
#10
def check(a):
    lst = []
    for i in a:
        if i in lst:
            return False
        else:
            lst.append(i)
    return True

try:
    a = list(map(int,input().split()))
    if check(a):
        a.pop(0)
        print(sum(a))
    else:
        print("Mang khong hop le")
except:
    print("Mang khong hop le")
```
