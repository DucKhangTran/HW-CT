```python
#3
def move_left(a, k):
    while k > 0:
        temp = a.pop(0)
        a.append(temp)
        k = k - 1
    return tuple(a)
a = tuple(map(int, input().split()))
k = int(input())
a = list(a)
print(move_left(a,k))
```

```python
#4
def dic(a):
    d = {}
    lst1 = []
    for x in a:
        x = x.split(':')
        lst1.append(x)
    for item in lst1:
        key = item[0]
        value = item[1]
        if key not in d:
            d[key] = [value]
        else:
            d[key].append(value)
    return d

a = list(map(str, input().split()))
print(dic(a)) 


    
    
        
```

```python
#5
def num_sort(a):
    d = {}
   
    d['positives'] = 0
    d['negatives'] = 0
    d['zeros'] = 0
    for num in a:
        if num > 0:
            d['positives'] += 1
        elif num < 0:
            d['negatives'] += 1
        else:
            d['zeros'] += 1
    return d
a = list(map(int, input().split()))
print(num_sort(a))
```

```python
#8
def num_of_words(a):
    d = {}
    for word in a:
        if word not in d:
            d[word] = 1
        else:
            d[word] += 1
    return d
a = list(map(str, input().split()))
print(num_of_words(a))
```

```python
#9
def dic_combine(a,b):
    d = {}
    for item in a:
        if item in b:
            d[item] = a[item] + b[item]
        else:
            d[item] = a[item]
    for item in b:
        if item not in a:
            d[item] = b[item]
    sorted_dict = list(d.items())
    sorted_dict.sort()
    return dict(sorted_dict)

a = eval(input())
b = eval(input())
print(dic_combine(a,b))
```

```python
#12
def most_app(a):
    app = []
    for x in a:
        app.append(a.count(x))
    b = max(app)
    most_app_num = []
    for x in a:
        if a.count(x) == b:
            most_app_num.append(x)
    return min(most_app_num)

a = list(map(int, input().split()))
print(most_app(a))

```

```python
#14
def same_item(a, b):
    lst = []
    for x in a:
        if x in b and x not in lst:
            lst.append(x)
            
    return lst
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(same_item(a,b))
```

```python
#15
def loc_dict(a, b):
    d = {}
    loc = {}
    for i in range(0,len(a), 2):
        d[a[i]] = int(a[i +1])
    for item in d:
        if d[item] > b:
            loc[item] = d[item]
    return loc
a = list(map(str, input().split()))
b = int(input())
print(loc_dict(a,b))
```

```python
#17
n = int(input())
matrix = []
chinh = []
phu = []
c = 0
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
for j in range(len(matrix)):
    chinh.append(matrix[j][j])
    for k in range(len(row) - 1, -1 , -1):
        phu.append(matrix[j][k - c])
        c += 1
        break
        

print('Đường chéo chính:', *chinh)
print('Đường chéo phụ:', *phu)
```

```python
#18
def sum_colK(n, k):
    sum_K = 0
    for i in range(len(n)):
        sum_K += n[i][k-1]
    return sum_K

n = int(input())
k = int(input())
matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

print(sum_colK(matrix, k))
```

```python
#20
def idx(a, k):
    for i in range(len(a)):
        if a[i] == k:
            return i + 1
    return -1

a = list(map(int,input().split()))
k = int(input())
print(idx(a,k))
```

```python
#21
def anhxa(a):
    d = {}
    for i in range(len(a)):
        if a[i] not in d:
            d[a[i]] = [i]
        else:
            d[a[i]].append(i)
    for k , v in d.items():
        d[k] = tuple(v)
    return d
a = list(map(str, input().split()))
print(anhxa(a))
        
```
