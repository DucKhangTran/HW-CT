```python
#1
class cyclinder:
    def __init__(self, r,h):
        self.r = r
        self.h = h
    def area(self):
        return 3.14*(self.r**2)*2 + 2*3.14*self.r*self.h
    def volume(self):
        return 3.14*(self.r**2)*self.h

n = input().split()
p = cyclinder(int(n[0]), int(n[1]))
print('Diện tích bề mặt:', p.area())
print('Thể tích:', p.volume())
```

```python
#2
class Date:
    def __init__(self, d, m, y):
        self.d = d
        self.m = m
        self.y = y
    def leap_year(self):
        return self.y % 400 == 0 or (self.y % 4 == 0 and self.y % 100 !=0)
    def max_day(self):
        days_in_month = [0,31,28,31,30,31,30,31,31,30,31,30,31]
        if self.m == 2 and self.leap_year():
            return 29
        return days_in_month[self.m]
    def is_valid(self):
        if self.d <1 and self.d > self.max_day():
            return False
        if self.m < 1 and self.m > 12:
            return False
        if self.y < 1:
            return False
        return True
    def next_day(self):
        self.d += 1
        if self.d > self.max_day():
            self.d = 1
            self.m += 1
            if self.m > 12:
                self.m = 1
                self.y += 1
        return f"{self.d}/{self.m}/{self.y}"
n = input().split('/')
d, m, y = int(n[0]), int(n[1]), int(n[2])
my_date = Date(d,m,y)
if my_date.is_valid():
    print(my_date.next_day())
else:
    print('INVALID')
```

```python
#3
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def is_origin(self):
        if self.x == 0 and self.y == 0:
            return 'M là gốc tọa độ'
        elif self.x == 0 and self.y != 0:
            return 'M nằm trên trục tung'
        elif self.x != 0 and self.y == 0:
            return "M nằm trên trục hoành"
        else:
            return 'M không nằm trên trục tung và trục hoành'
    def distance(self):
        return round((self.x**2 + self.y**2)**(1/2), 2)

n = list(map(float, input().split()))
x,y = n[0], n[1]
M = Point(x,y)
if M.distance() == 0.00:
    print('M là gốc tọa độ')
else:
    print(M.is_origin())
    print('Khoảng cách từ M đến O:', M.distance())
        
```

```python
#4
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def add(self):
        return self.a + self.b
    def subtract(self):
        return self.a - self.b
    def product(self):
        return self.a * self.b
    def divide(self):
        return round(self.a / self.b, 3)
    def power(self):
        return self.a** self.b
    def mod(self):
        return self.a % self.b
    def set_numbers(self):
        self.a = x
        self.b = y

set_num = True
while set_num == True:
    n = list(map(str, input().split()))
    a,b = int(n[0]), int(n[2])
    c = n[1]
    nums = Calculator(a,b)
    if c == '+':
        print(nums.add())
    elif c == '-':
        print(nums.subtract())
    elif c == 'x':
        print(nums.product())
    elif c == ':' or c == '/':
        print(nums.divide()) 
    elif c == '**':
        print(nums.power())
    elif c == '%':
        print(nums.mod())
    change_num = input('Đổi số?')
    if change_num == 'YES' or change_num == 'yes' or change_num == 'Yes':
        set_num = True
    else:
        set_num = False

```

```python
#5
class ShoppingCart:
    def __init__(self):
        self.product = []
    def add(self):
        name = input("Product's name")
        price = float(input("Product's price"))
        item = {'name': name, 'price': price}
        self.product.append(item)
        return self.product
    def remove(self):
        name2remove = input('Item to be removed:')
        for item in self.product:
            if item['name'] == name2remove:
                self.product.remove(item)
                return f'removed {name2remove}'
        return f'{name2remove} does not exist'
    def check_empty(self):
        if len(self.product) == 0:
            return 'Empty'
        else:
            return 'not empty'
    def total_price(self):
        total_price = 0
        for item in self.product:
            total_price += item['price']
        return total_price
    def show_product(self):
        for item in self.product:
            print( f'{item['name']} - {item['price']}')
    def delete_cart(self):
        self.product = []


shopping = ShoppingCart()
print(' -----Shopping Menu------')
print('2. Add items')
print('3. Remove item')
print('4. Check cart')
print('5. Total price')
print('6. Show cart')
print('7. Delete cart')
while True:
    choice = int(input('My choice is (2-7):'))
    if choice ==  2:
        shopping.add()
    elif choice == 3:
        print(shopping.remove())
    elif choice == 4:
        print(shopping.check_empty())
    elif choice == 5:
        print(shopping.total_price())
    elif choice == 6:
        shopping.show_product()
    elif choice == 7:
        shopping.delete_cart()
    
    ask = input('Do you want to exit?')
    if ask.lower() == 'yes':
        shopping.show_product()
        break
        

        
        
```

```python
#6
class Rectangle:
    def __init__(self,a,b):
        self.a = a
        self.b = b
       
    def area(self):
        return (self.a*self.b,2)
    def perimeter(self):
        return (self.a + self.b)*2
    def scale(self,k):
        self.a = k * self.a
        self.b = k * self.b

n = list(map(float, input().split()))
a, b,c = n[0], n[1], n[2]
rec = Rectangle(a,b)
rec.scale(c)
print(f'{rec.area():.2f} {rec.perimeter():.2f}')

        
```

```python
#7
import math as m
class Fraction:
    def __init__(self, num , den):
        self.num = num
        self.den = den
        if self.den < 0:
            self.num = -self.num
            self.den = -self.den # mẫu > 0
        gcd = m.gcd(self.num, self.den)
        self.num = int(self.num/ gcd )
        self.den = int(self.den/ gcd)
    def add(self, other):
        new_num = self.num * other.den + self.den * other.num 
        new_den = self.den*other.den
        return Fraction(new_num, new_den)
    def sub(self, other):
        new_num = (self.num * other.den - self.den * other.num) 
        new_den = (self.den*other.den)
        return Fraction(new_num, new_den)
    def mul(self, other):
        new_num = self.num*other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
    def div(self, other):
        new_num = self.num* other.den
        new_den = self.den*other.num
        return Fraction(new_num,new_den)
    def __str__(self):
        return f'{self.num}/{self.den}'

frac = input().split()
a,b, op, c, d = frac[0], frac[1], frac[2], frac[3], frac[4]
f1 = Fraction(int(a),int(b))
f2 = Fraction(int(c),int(d))
if op == '+':
    print(f1.add(f2))
elif op == '-':
    print(f1.sub(f2))
elif op == '*':
    print(f1.mul(f2))
elif op == '/':
    print(f1.div(f2))


```

```python
#8
class BankAccount:
    def __init__(self,owner, balance):
        self.owner = owner
        self.balance = balance
    def deposit(self, x):
        self.balance += x
        return self.balance
    def withdraw(self, x):
        if x > self.balance:
            pass
        else:
            self.balance -= x
        return self.balance
    def __str__(self):
        return f'Your balance is {self.balance:.2f}'
    
owner_in4 = input().split()
owner, balance = owner_in4[0], float(owner_in4[1])
account = BankAccount(owner, balance)
n = int(input('Number of commands'))
for i in range(n):
    command_amount = input().split()
    if command_amount[0] == 'DEPOSIT':
        account.deposit(float(command_amount[1]))
    elif command_amount[0] == 'WITHDRAW':
        account.withdraw(float(command_amount[1]))

print(account)
   


    


```

```python
#9
class Student:
    def __init__(self, name):
        self.average = []
        self.name = name
    def avg(self,x):
        for i in range(x):
            subject = input().split()
            self.average.append(float(subject[1]))
        return f'{(sum(self.average)/len(self.average)):.2f}'
    def rank(self):
        avg =  sum(self.average)/len(self.average)
        if avg >= 8:
            return 'Excellent'
        elif 6.5 <= avg < 8:
            return 'Good'
        elif 5 < avg < 6.5:
            return 'Average'
        else:
            return 'Poor'
   
name = input()
nums_sub = int(input())
average = Student(name)
print(f'{name} {average.avg(nums_sub)} {average.rank()}')
```
