```python
#1
def Merge(a,b):
    c=a+b
    n = len(c)
    for i in range(len(c)):
        for j in range(n-i-1):
            if c[j]>c[j+1]:
                c[j], c[j+1]=c[j+1], c[j]
    return c
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(*Merge(a,b))
```

```python
#2
def count_less_equal(matrix, n, m, target):
    count = 0
    row = n - 1
    col = 0
    while row >= 0 and col < m:
        if matrix[row][col] <= target:
            count += (row + 1)
            col += 1
        else:
            row -= 1
    return count


def find_kth_smallest(matrix, k):
    n = len(matrix)
    m = len(matrix[0])

    low = matrix[0][0]
    high = matrix[n - 1][m - 1]

    while low < high:
        mid = (low + high) // 2
        if count_less_equal(matrix, n, m, mid) < k:
            low = mid + 1
        else:
            high = mid

    return low

matrix = []
k = 0

while True:
    line = list(map(int, input().split()))
    if len(line) == 1:
        k = line[0]
        break
    matrix.append(line)

print(find_kth_smallest(matrix, k))
```

```python
#4
import sys

def solve():
    input_data = sys.stdin.read().split()
    if not input_data:
        return
    iterator = iter(input_data)
    try:
        q = int(next(iterator))
    except StopIteration:
        return

    for _ in range(q):
        try:
            a = next(iterator)
            b = next(iterator)
        except StopIteration:
            break

        n = len(a)
        m = len(b)
        dp = [[False] * (m + 1) for _ in range(n + 1)]
        dp[0][0] = True
        for i in range(1, n + 1):
            if a[i-1].islower():
                dp[i][0] = dp[i-1][0]
            else:
                dp[i][0] = False 
        for i in range(1, n + 1):
            for j in range(1, m + 1):
                char_a = a[i-1]
                char_b = b[j-1]

                if char_a.isupper():
                   
                    if char_a == char_b:
                        dp[i][j] = dp[i-1][j-1]
                    else:
                        dp[i][j] = False
                else:
                    option_delete = dp[i-1][j]
                    option_match = False
                    if char_a.upper() == char_b:
                        option_match = dp[i-1][j-1]
                    
                    dp[i][j] = option_delete or option_match

        if dp[n][m]:
            print("YES")
        else:
            print("NO")

if __name__ == "__main__":
    solve()
```

```python
#5
def solve(lst1, lst2):
  n1 = len(lst1)
  n2 = len(lst2)
  if n1 == 0:
    return True
  if n1>n2:
    return False
  for i in range(n2-n1+1):
    if lst1 == lst2[i:i+n1]:
      return True
  return False
nums1 = list(map(int, input().split()))
nums2 = list(map(int, input().split()))
print(solve(nums1, nums2))
```

```python
#6
def hoanvi(arr, n):
  i = n - 2
  while arr[i+1]<=arr[i]:
    i-=1
  j = n - 1
  while(arr[j] < arr[i]):
    j -= 1
  arr[i], arr[j] = arr[j], arr[i]
  ans = arr[0:i+1]+list(reversed(arr[i+1:]))
  return ans
n = int(input())
lst = list(map(int, input().split()))
print(*hoanvi(lst, n))
```

```python
#7
def canPlaceFlowers(flowerbed, k):
  count = 0
  n = len(flowerbed)
  for i in range(n):
    if flowerbed[i] == 0 and (i == 0 or flowerbed[i-1] == 0) and (i == n-1 or flowerbed[i+1] == 0):
      flowerbed[i] = 1
      count += 1
    if count >= k:
      return True
  return False

lst = list(map(int, input().split()))
k = int(input())
print(canPlaceFlowers(lst, k))
```

```python
#8
def is_path_crossing(moves):
  x, y = 0, 0
  di_qua = {(0, 0)}
  for move in moves:
    if move == 'R':
      x+=1
    if move == 'L':
      x-=1
    if move == 'U':
      y+=1
    if move == 'D':
      y-=1
    if (x, y) in di_qua:
      return True
    di_qua.add((x, y))
  return False
moves = input()
print(is_path_crossing(moves))
```

```python
#9
def Median(nums):
  if nums[0] > nums[1]:
    nums[0], nums[1] = nums[1], nums[0]
  if nums[2] > nums[3]:
    nums[2], nums[3] = nums[3], nums[2]
  if nums[0] > nums[2]:
    nums[0], nums[2] = nums[2], nums[0]
    nums[1], nums[3] = nums[3], nums[1]
  if nums[1] > nums[4]:
    nums[1], nums[4] = nums[4], nums[1]
  if nums[1] > nums[2]:
    nums[1], nums[2] = nums[2], nums[1]
    nums[3], nums[4] = nums[4], nums[3]
  return nums[2] if nums[2] < nums[4] else nums[4]
nums = list(map(int, input().split()))
print(Median(nums))
```

```python
#10
def findRadius(houses, heaters):
  for i in range(len(houses)):
    min_dis = -1
    for j in range (len(heaters)):
      dis = abs(houses[i] - heaters[j])
      if min_dis == -1 or dis < min_dis:
        min_dis = dis
    houses[i] = min_dis
  return max(houses)
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(findRadius(houses, heaters))
```
