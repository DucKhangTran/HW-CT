```python
#W4A8
def reversenum(n):
    x = 0
    while n > 0:
        x = x*10 + n % 10
        n = n //10
    return x
def palindrome(n):
    return n == reversenum(n)

n = int(input())
count = 0
while palindrome(n) == False:
    n += reversenum(n)
    count += 1
print(count, n)
```

```python
#32
def password(n):
    count1 = 0
    count2 = 0
    count3 = 0
    count4 = 0
    
    if len(n) < 0:
        return False
    else:
        for i in range(len(n)):
            if n[i].isupper():
                count1 += 1
            elif n[i].islower():
                count2 += 1
            elif n[i].isnumeric():
                count3 += 1
            elif not n[i].isalpha() and not n[i].isdigit():
                count4 += 1
    if count1 > 0 and count2 > 0 and count3 > 0 and count4 > 0:
        return True
    else:
        return False

n = input()
if password(n) == True:
    print('Strong password')
else:
    print('Weak password')

    

```

```python
#34
def chuoi(a,b):
    for i in range(len(a)):
        if i == 0:
            if b == a[0: len(b)]:
                return a[len(b): len(a)]
        else:
            if b == a[i:i + len(b)]:
                return (a[0: i] + a[i + len(b) + 1: len(a)])
a = input()
b = input()
print(chuoi(a,b))
```
