```python
#1
def binary_search(arr, value):
    left , right = 0 , len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == value:
            return mid
        elif arr[mid] > value:
            right = mid - 1
        else:
            left = mid + 1
    return - 1
arr = list(map(int, input().split()))
value = int(input())
print(binary_search(arr, value))
```

```python
#3 bubble
def buble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n - i -1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
n = list(map(int,input().split()))
print(*buble_sort(n))
```

```python
#3 selection
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx] , arr[i]
    return arr
n = list(map(int, input().split()))
print(*selection_sort(n))
```

```python
#4
def most_app(arr):
    d = {}
    max = 0
    for i in arr:
        if i not in d:
            d[i] = 1
        else:
            d[i] += 1
        if d[i] > max:
            max = d[i]
    for item in d:
        if d[item] == max:
            return f"{item} xuat hien nhieu nhat, som nhat, {max} lan"

n = list(map(int, input().split()))
print(most_app(n))
```

```python
#5
def pair(lst, a):
    x = []
    for i in range(len(lst)):
        for j in range(i + 1, len(lst)):
            if lst[i] + lst[j] == a:
                x.append([lst[i], lst[j]])
    return len(x)

a = list(map(int, input().split()))
b = int(input())
print(pair(a,b))
```

```python
#8
def solve(intervals, queries):
    ket_qua = []
    for q in queries:
        min_len = -1  
        for val in intervals:
            l = val[0]
            r = val[1]
            if l <= q and q <= r:
                do_dai = r - l + 1
                if min_len == -1 or do_dai < min_len:
                    min_len = do_dai
        ket_qua.append(min_len)
    return ket_qua

intervals = eval(input())
queries = eval(input())

print(solve(intervals, queries))
```

```python
#9
def compare(a):
    nums = []
    count = 0
    for i in range(len(a)):
        for j in range(i, len(a)):
            if a[i] > a[j]:
                count += 1
        nums.append(count)
        count = 0
    return nums

a = eval(input())
print(compare(a))
```

```python
#10
import bisect
nums = eval(input())
m = int(input())
ans = 0
curr = 0
prefixes = [0]
for x in nums:
    curr = (curr + x) % m
    ans = max(ans, curr)
    idx = bisect.bisect_right(prefixes, curr)
    if idx < len(prefixes):
        ans = max(ans, curr - prefixes[idx] + m)
    bisect.insort(prefixes, curr)

print(ans)
```

```python
#11
import string
def count_words(a,b):
    count = 0
    if len(b) < 2:
        for word in a:
            if word == b[0]:
                count += 1
    else:
        for i in range(0, len(a)):
            if a[i:i +len(b)] == b:
                count += 1
    return count

a = input()
for dau_cau in string.punctuation:
    a = a.replace(dau_cau, ' ')
temp = a.split()
b = input().split()
print(count_words(temp,b))


```

```python
#12
def fastest_finger(a):
    d = {}
    d['A'] = a[0]
    d['B'] = a[1]
    d['C'] = a[2]
    d['D'] = a[3]
    temp_lst = []
    for k, v in d.items():
        temp_lst.append([v,k])
    temp_lst.sort()
    ans = []
    for i in range(len(temp_lst)):
        ans.append(temp_lst[i][1])
    return ans

a = []
for i in range(4):
    a.append(input().lower())
print(*fastest_finger(a))


```

```python
#13
s = input()
if '[' in s:
    nums = eval(s)
else:
    nums = list(map(int, s.split()))
counts = {}
for x in nums:
    if x in counts:
        counts[x] += 1
    else:
        counts[x] = 1
ans = 0
for x in counts:
    if (x + 1) in counts:
        tong = counts[x] + counts[x + 1]
        if tong > ans:
            ans = tong
print(ans)
```
